## create vector and list

```python
from Bio import SeqIO
```

```python
%%timeit 
ids = []
with gzip.open(plate1, "rt") as handle:
    records = SeqIO.parse(handle, "fastq")
    for rec in records:
        umi, cb = extract_umi_cb(str(rec.id))
        seq = rec.seq
        id_item = (seq, cb, umi)
        ids.append(id_item)
df = pd.DataFrame(ids, columns=['seq', 'cb','umi'])
dedup_df = df.drop_duplicates().reset_index(drop=True)
count_df = dedup_df.groupby(['seq', 'cb']).count()
seqcell_table = count_df.unstack(fill_value=0)
```
